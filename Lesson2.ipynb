```python
df1 = df
```

```python
df1 = df1.replace({'label':{'negative':1, 'positive':0}})
```

```python
x_train, x_test, y_train, y_test = train_test_split(df1.text, df1.label)
```

```python
from sklearn.linear_model import LogisticRegression # можно заменить на любимый классификатор
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from nltk import ngrams
import nltk
from nltk import collocations 
from nltk.tokenize import word_tokenize, wordpunct_tokenize
from nltk.corpus import stopwords
nltk.download('genesis')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from string import punctuation
```

```python
stop_punct = stopwords.words('english') + list(punctuation)
tokens = [token for sense in df1.text for token in word_tokenize(sense) if token not in punctuation]
```

```python
from collections import Counter
```

```python
freq = Counter(tokens)
freq_sort = sorted(freq.items(), key=lambda x: -x[1])
max([el[1] for el in list(freq_sort)]), min([el[1] for el in list(freq_sort)])
```

```python
high_tokens = set()
j = 10000
for i in freq_sort:
  if i[1] > j:
    high_tokens.add(i[0])
print(len(high_tokens))
```

```python
med_tokens = set()
t = 1000
for i in freq_sort:
  if i[1] < j and i[1] > t:
    med_tokens.add(i[0])
print(len(med_tokens))
```

```python
low_tokens = set()
for i in freq_sort:
  if i[1] < t:
    low_tokens.add(i[0])
print(len(low_tokens))
```

```python
red = stop_punct + list(high_tokens)
```

```python
vec = CountVectorizer(ngram_range=(1, 1), tokenizer=word_tokenize, stop_words=red)
bow = vec.fit_transform(x_train)
clf = LogisticRegression(random_state=42)
clf.fit(bow, y_train)
pred = clf.predict(vec.transform(x_test))
print(classification_report(pred, y_test))
```

```python
red1 = stop_punct + list(med_tokens)
```

```python
vec = CountVectorizer(ngram_range=(1, 1), tokenizer=word_tokenize, stop_words=red1)
bow = vec.fit_transform(x_train)
clf = LogisticRegression(random_state=42)
clf.fit(bow, y_train)
pred = clf.predict(vec.transform(x_test))
print(classification_report(pred, y_test))
```

```python
red2 = stop_punct + list(low_tokens)
```

```python
vec = CountVectorizer(ngram_range=(1, 1), tokenizer=word_tokenize, stop_words=red2)
bow = vec.fit_transform(x_train)
clf = LogisticRegression(random_state=42)
clf.fit(bow, y_train)
pred = clf.predict(vec.transform(x_test))
print(classification_report(pred, y_test))
```
